## (ii) Numerical ##

### Number of screens ###

```python
sns.distplot(data['Number_of_Screens'])
plt.title('Distribution of no. of screens')
```

```python
sns.distplot(data['Revenue(INR)'], color='orangered')
plt.title('Revenue distribution')
```

```python
sns.distplot(data['Budget(INR)'], color ='Black')
plt.title('Distribution of Budget')
plt.show()
```

```python
data["Number_of_Screens"].skew()
```

The data is highly skewed. Most of the films have got close to zero screens.

```python
top_movies_by_screens = data.sort_values(by='Number_of_Screens', ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x= 'Movie_Name', y= 'Number_of_Screens', data=top_movies_by_screens)
plt.xticks(rotation=45)
plt.show()
```

```python
top_movies_by_revenue = data.sort_values(by='Revenue(INR)', ascending=False).head(20)
plt.figure(figsize=(15,6))
sns.barplot(x= 'Movie_Name', y= 'Revenue(INR)', data=top_movies_by_revenue)
plt.xticks(rotation=45)
plt.show()
```

## Bivariate Analysis ##

### Numerical Numerical ###

Let us plot scatterplots to see the correlation between numerical variables.

```python
sns.scatterplot(data = data,x="Number_of_Screens",y="Budget(INR)")
plt.show()
```

```python
sns.scatterplot(data = data,x="Number_of_Screens",y="Revenue(INR)")
plt.show()
```

```python
sns.scatterplot(data = data,x="Budget(INR)",y="Revenue(INR)")
plt.show()
```

```python
#Correlation matrix
corr = data.corr()
print(corr)
```

```python
# Correlation heatmap among numeric variables
plt.figure(figsize=(8,6))
corr = data.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Numeric Features')
plt.tight_layout()
plt.show()
```

The scatterplots and correlation heatmap illustrate that all three numerical attributes have high positive correlation.
1. Revenue and number of screens are correlated and if number of screens are increased, revenue can increase.
2. Similarly Revenue and Budget are correlated as big budget are mostly earning big revenues.
3. Number of screens and budget are also correlated but they have least correlation as budget may also depend on other factors like, lead stars, etc.

### (ii) Categorical and numerical

```python
plt.figure(figsize=(10,6))
sns.barplot(data=data,x="Genre",y="Revenue(INR)")
plt.xticks(rotation = 45)
plt.show()
```

This barplot shows that masala movies have got highest average revenue. Then come the action movies but with significant difference.

Adult movies have the least average revenue.

```python
# Box plot to assess Revenue distribution across Genres
plt.figure(figsize=(10,6))
sns.boxplot(data=data, x='Genre', y='Revenue(INR)')
plt.title('Revenue Distribution by Genre')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Grouped bar plot: Average Revenue segmented by 'Whether_Remake' and 'Whether_Franchise'
plt.figure(figsize=(9,5))
grouped = data.groupby(['Whether_Remake', 'Whether_Franchise'])['Revenue(INR)'].mean().reset_index()
sns.barplot(data=grouped, x='Whether_Remake', y='Revenue(INR)', hue='Whether_Franchise')
plt.title('Average Revenue by Remake and Franchise status')
plt.tight_layout()
plt.show()
```

```python
sns.heatmap(pd.crosstab(top_movies_by_revenue['Release_Period'],top_movies_by_revenue['Revenue(INR)']))
plt.show()
```
